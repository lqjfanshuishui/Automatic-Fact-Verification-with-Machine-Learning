## Creating the index for each documents
Using pylucene to index the document names in the wiki files

```python
import sys, os, lucene, threading, time
from datetime import datetime

from java.nio.file import Paths
from org.apache.lucene import analysis, document, index, queryparser, search, store, util
from org.apache.lucene.document import Document, Field, FieldType

from org.apache.lucene.store import SimpleFSDirectory
assert lucene.getVMEnv() or lucene.initVM()

# # # lucene # # #

analyzer = analysis.standard.StandardAnalyzer()

# Store the index in memory:

store = SimpleFSDirectory(Paths.get('/Users/qijieli/Desktop/web_project_code/doc_name_index'))

#define feilds
t1 = FieldType()
t1.setStored(True)
t1.setTokenized(True)
t1.setIndexOptions(index.IndexOptions.DOCS_AND_FREQS)

config = index.IndexWriterConfig(analyzer)
iwriter = index.IndexWriter(store, config)

prefix = "./wiki-pages-text/wiki-"
suffix  = ".txt"

for i in range(109):
    start = time.perf_counter()
    if i <9:
        wikiNum = "00"+str(i+1)
    elif i<99:
        wikiNum = "0"+str(i+1)
    else:
        wikiNum = str(i+1)

    filename = prefix+wikiNum+suffix
    print("wiki-"+wikiNum)
    
    doc_list = []
    content = ""
    for line in open(filename):
        doc_title,sent_id,sent_content = line.split(" ",2)
        
        if sent_id.isalpha():
            
            pass
        else:
            if doc_title not in doc_list:
                doc_list.append(doc_title)
                doc = document.Document()
                content = " ".join(doc_title.split("_"))
                doctitle = doc_title
#                 print(content)
#                 print(doctitle)
                doc.add(document.Field("wiki", str(i), document.TextField.TYPE_STORED))
                doc.add(document.Field("docID", doctitle, document.TextField.TYPE_STORED))
                doc.add(document.Field("content", content, t1))
                iwriter.addDocument(doc)
                
    
    elapsed = (time.perf_counter() - start)
    print("Time used:",elapsed)






iwriter.close()
store.close()


```
